```python
# processing

# vieets lai, phan loai opening & closing

from datetime import datetime
def convert_time_open(value):
    if pd.isnull(value) or value=='':
        return None
    hour = int(value//100)
    minute = int(value % 100)
    date = datetime(2024,1,1,hour,minute)
    return date.strftime('%Y-%m-%d %H:%M:%S')
def convert_time_close(value):
    if pd.isnull(value) or value=='':
        return None
    if value<=500:
        hour = int(value//100)
        minute = int(value % 100)
        date = datetime(2024,1,2,hour,minute)
        return date.strftime('%Y-%m-%d %H:%M:%S')
    hour = int(value//100)
    minute = int(value % 100)
    date = datetime(2024,1,1,hour,minute)
    return date.strftime('%Y-%m-%d %H:%M:%S')
open_col = [col for col in a.columns if 'day_open' in col]
close_col = [col for col in a.columns if 'day_close' in col]
for col in open_col:
    a[col] = pd.to_numeric(a[col],errors='coerce').apply(lambda z: convert_time_open(z))
for col in close_col:
    a[col] = pd.to_numeric(a[col],errors='coerce').apply(lambda z: convert_time_close(z))
    
    
```
